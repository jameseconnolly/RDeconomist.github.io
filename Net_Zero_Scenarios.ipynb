```python
import pandas as pd
```

```python
df = pd.read_csv('long_target_emissions.csv')
```

```python
# // Cleaning and organising df
df.drop(df.index[6:10], inplace=True)
df = df.rename(columns={'Unnamed: 0': 'Scenario'})
df.drop('2100', axis=1, inplace=True)
df
```

```python
# // Converting from wide to tidy format
df1=df.set_index(['Scenario']).stack().reset_index()
```

```python
df1 = df1.rename(columns={'level_1': 'Year',0:'Global GHG emissions/year'})
```
